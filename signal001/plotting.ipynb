```python
import numpy as np

def signal_load():
	import csv
	f = open('/Users/tmarkovich/Dropbox/Projects/CSComparisonPaper/signals/signal001/signal001.csv', 'rb')
	reader = csv.reader(f)
	signal = []
	for row in reader:
	    signal.append(row)
	f.close()
	signal = np.array(signal).astype(np.float)
	time = signal[:,0]
	signal = np.squeeze(signal[:,1])
	return time, signal
```
