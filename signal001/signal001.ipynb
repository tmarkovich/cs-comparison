```python
import numpy as np
import imp
import csv

cs_utils = imp.load_source('cs_utils',"/Users/tmarkovich/Dropbox/Projects/cslib2/cs_utils.py")
Astruct = imp.load_source('Astruct', '/Users/tmarkovich/Dropbox/Projects/cslib2/Astruct.py')
twist = imp.load_source('twist',"/Users/tmarkovich/Dropbox/Projects/cslib2/TwIST_Solver.py")

debug = False
```

```python
f = open('/Users/tmarkovich/Dropbox/Projects/CSComparisonPaper/signals/signal001/signal001.csv', 'rb')
reader = csv.reader(f)
signal = []
for row in reader:
    signal.append(row)
f.close()
signal = np.array(signal).astype(np.float)
time = signal[:,0]
signal = np.squeeze(signal[:,1])
cs_utils.thomas_ploting(time, [signal], 
                        title='Original Signal to Reproduce',
                        legend_label=['Original Signal'],
                        xlabel='Time [s]',
                        ylabel='Signal [Arb]')
```

```python
def step_function(x, inflection):
    return np.sign(time - inflection)


def complementary_step_function(x, inflection):
    return np.sign(inflection - time)


def sine(x, freq):
    return np.sin(x*freq)

# Plotting to check
if debug:
    cs_utils.thomas_ploting(time, [step_function(time, 0.3)], 
                            title='Step Function with Inflection at 0.3s',
                            legend_label=['Inflection at 0.3s'],
                            xlabel='Time [s]',
                            ylabel='Signal [Arb]')


    cs_utils.thomas_ploting(time, [complementary_step_function(time, 0.3)], 
                            title='Complimentary Step Function with Inflection at 0.3s',
                            legend_label=['Inflection at 0.3s'],
                            xlabel='Time [s]',
                            ylabel='Signal [Arb]')


    cs_utils.thomas_ploting(time, [sine(time, 30)], 
                            title='Sine with Frequency 30 Hz',
                            legend_label=['Frequency 30 Hz'],
                            xlabel='Time [s]',
                            ylabel='Signal [Arb]')
```

```python
def matrix_maker_signal001(x, minfreq = 0, maxfreq = 400, dw = 0.1, debug = False):
    # Creates the step function structure
    A_step = Astruct.Astruct(step_function, 
                  x, 
                  np.linspace(0, 1, 101))
    A_step.matrix = cs_utils.matrix_maker(A_step)
    
    # Creates the complimentary step function structure
    A_comp_step = Astruct.Astruct(complementary_step_function, 
                      x, 
                      np.linspace(0, 1, 101))
    A_comp_step.matrix = cs_utils.matrix_maker(A_comp_step)

    # Creates the sine function structure
    freq = np.arange(minfreq, maxfreq+dw, dw)
    A_sine = Astruct.Astruct(sine, 
                  x, 
                  freq)
    A_sine.matrix = cs_utils.matrix_maker(A_sine)
    A = np.hstack((A_sine.matrix, A_step.matrix, A_comp_step.matrix))

    # Plots the three functions to check it all
    if debug:
        cs_utils.thomas_ploting(time, [A_step.matrix[:,23], A_comp_step.matrix[:,23], A_sine.matrix[:,203]], 
                                title='Checking the matrix generation',
                                legend_label=['Step Function', 'Complimentary Step Function', 'Sine Function'],
                                xlabel='Time [s]',
                                ylabel='Signal [Arb]')
    return A_step, A_comp_step, A_sine, A


def reproduction(time, signal, lambdas, A_step, A_comp_step, A_sine, A, plotting = False):
    reproduced = np.zeros(shape=np.shape(time))
    reproduced += cs_utils.reproduce_series(A_sine, lambdas_debias[0:len(A_sine.params)])

    offset = len(A_sine.params)
    reproduced += cs_utils.reproduce_series(A_step, lambdas_debias[offset:offset+len(A_step.params)])

    offset += len(A_step.params)
    reproduced += cs_utils.reproduce_series(A_comp_step, lambdas_debias[offset:offset+len(A_comp_step.params)])
    l2Error = np.sum((signal - reproduced)**2) / np.sum((signal)**2)
    if(plotting):
        cs_utils.thomas_ploting(time, [signal, reproduced], 
                                title='Original Signal vs Reproduction',
                                legend_label=['Original Signal', 'Reproduced'],
                                xlabel='Time [s]',
                                ylabel='Signal [Arb]')
        print 'L2 integrated error = ', l2Error
    return reproduced, l2Error
```

```python
A_step, A_comp_step, A_sine, A = matrix_maker_signal001(time, debug = True)
twist.tolA = 1e-7
twist.tolD = 1e-9
twist.verbose = False
lambdas, lambdas_debias, objective, times, debias_start, max_svd = twist.solve(signal, A)
l2Error, reproduction = reproduction(time, signal, lambdas, A_step, A_comp_step, A_sine, A, plotting = True)
```
