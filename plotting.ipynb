# Signal 1

This is the plotting piece of the notebook. You should define where the signal is as well as the directory where all the reproduced signals are.

From here, you'll simply pick an n which indexes how much of the signal is included (each n is 64 more time points).

```python
def signal_load():
	import csv
	f = open('/Users/tmarkovich/Dropbox/Projects/CSComparisonPaper/signals/signal001/signal001.csv', 'rb')
	reader = csv.reader(f)
	signal = []
	for row in reader:
	    signal.append(row)
	f.close()
	signal = np.array(signal).astype(np.float)
	time = signal[:,0]
	signal = np.squeeze(signal[:,1])
	return time, signal
dir = '/Users/tmarkovich/Dropbox/Projects/CSComparisonPaper/signals/signal001/'

errors_CS = np.load(dir+'errors_CS.npy')
errors_SR = np.load(dir+'errors_SR.npy')
errors_HI = np.load(dir+'errors_HI.npy')
figure(figsize=(14, 10))
plot(log10(errors_CS[1]),linewidth='5')
plot(log10(errors_SR[1]),linewidth='5')
plot(log10(errors_HI[1]),linewidth='5')
legend(('CS','SR','HI'))

reproduced_CS = np.load(dir+'reproduced_CS.npy')
reproduced_SR = np.load(dir+'reproduced_SR.npy')
reproduced_HI = np.load(dir+'reproduced_HI.npy')

time, signal = signal_load()
n = 4

figure(figsize=(14, 10))
# plot(time, reproduced_HI[:,n])
plot(time,signal, linewidth='5')
legend(('Exact Signal','Compressed Sensing','Super Resolution','Harmonic Inversion'))
```
