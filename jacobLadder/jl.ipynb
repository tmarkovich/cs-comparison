# Signal Generation

This simply generates the signal couplet that Vlad wanted to generate in the paper. Note, I've taken on the real part (this shouldn't be an issue) and fixed a sign error that he had.

```python
import numpy as np
import imp

time = np.linspace(0, 999, num=1000)
signal = np.zeros(shape=(np.shape(time)))
for n in xrange(0, len(signal)):
    sum = 0 + 0*1j
    for m in xrange(0, 49):
        arg = 2*np.pi*0.9**m*2*10**-4
        if n == 1:
            print arg
        signal[n]  += (np.cos(arg*2500*n)+np.cos(arg*2487.5*n)+np.cos(arg*2475.0*n))*np.exp(-arg*n)
    signal[n] *= 81.92

figure(figsize=(14, 10))
plot(np.real(signal), linewidth='5')

np.savetxt('time.csv', time, delimiter=",")
np.savetxt('signal.csv', signal, delimiter=",")

time_range = [1, 2, 4, 5, 8, 10, 20, 25, 40, 50, 100]
```

# Compressed Sensing

```python
dunits = imp.load_source('units',"/Users/tmarkovich/Dropbox/Projects/csfree/units.py")
sr = imp.load_source('sr',"/Users/tmarkovich/Dropbox/Projects/cslibrary/cs.py")

def matrixGeneration(time):
    sr.allocateFreqs()
    sr.matrixBuilderDL(time)

sr.method = 'DL'
sr.maxiterm = 4000
dt = np.diff(time)[0]

asdf = np.zeros((49,3))
asdfGamma = np.zeros((50,))

for m in xrange(0, 49):
    arg = 2*np.pi*0.9**m*2*10**-4
    asdfGamma[m] = arg
    asdf[m,0] = arg*2500
    asdf[m,1] = arg*2487.5
    asdf[m,2] = arg*2475.0

sr.omegaGrid = np.reshape(asdf, (49*3,))
sr.gammaGrid = asdfGamma
sr.Verbosity = False

j = 0
tIncluded = 20
reproducedCS = np.zeros((len(time), 1000/20))
errorsCS = np.zeros((3,1000/20))

indices = np.arange(time.size)
np.random.shuffle(indices)

while tIncluded <= 1000:
    print "j =", j, " and tIncluded =", tIncluded
    idx = np.sort(indices[0:tIncluded])
    timeSlice = np.sort(time[idx])
    matrixGeneration(timeSlice)
    lambdas = sr.l1min(signal[idx])
    DOS, reproducedCS[:,j] = sr.constructSeriesDL(time, sr.freq, lambdas)
    
    errorsCS[0,j] = np.linalg.norm(reproducedCS[:,j]-signal , ord=1)/np.linalg.norm(signal, ord=1) 
    errorsCS[1,j] = np.linalg.norm(reproducedCS[:,j]-signal , ord=2)/np.linalg.norm(signal, ord=2) 
    errorsCS[2,j] = np.linalg.norm(reproducedCS[:,j]-signal , ord=np.inf)/np.linalg.norm(signal, ord=np.inf)
    
    tIncluded += 20
    j = j + 1
    
np.save('reproduced_CS.npy', reproducedCS)
np.save('errors_CS.npy', errorsCS)
```

# Filter Diagonalization

```python
harminv = imp.load_source('harminv',"/Users/tmarkovich/Dropbox/Projects/cslibrary/harminv.py")

dt = 1
n = len(signal)
nf = 6000
harminv.dens = 2.0
harminv.NF_MAX = 10000
harminv.NF_MIN = 6000
j = 0

reproducedHI = np.zeros((len(time), 1000/20), dtype=np.complex128)
errorsHI = np.zeros((3,1000/20))
tIncluded = 20

while tIncluded <= 1000:
    print "j =", j, " and tIncluded =", tIncluded
    n = len(signal[0:tIncluded])
    harminv.dataCreate(n, signal[0:tIncluded], 0.0*dt, 1.5*np.pi, 500)
    harminv.solve_once()
    harminv.compute_amps()
    reproducedHI[:,j] = harminv.reproduce(time)
    errorsHI[0, j] = np.linalg.norm(reproducedHI[:,j]-signal , ord=1)/np.linalg.norm(signal, ord=1) 
    errorsHI[1, j] = np.linalg.norm(reproducedHI[:,j]-signal , ord=2)/np.linalg.norm(signal, ord=2) 
    errorsHI[2, j] = np.linalg.norm(reproducedHI[:,j]-signal , ord=np.inf)/np.linalg.norm(signal, ord=np.inf)
    filename = 'parameters'+str(j)+'.txt'
    np.save(filename, harminv.data.u)
    tIncluded += 20
    j += 1

np.save('reproduced_HI.npy', reproducedHI)
np.save('errors_HI.npy', errorsHI)
```

# Filter Diagonalization Debugging

```python
harminv = imp.load_source('harminv',"/Users/tmarkovich/Dropbox/Projects/cslibrary/harminv.py")

dt = 1
n = len(signal)
nf = 6000
harminv.dens = 2.0
harminv.NF_MAX = 10000
harminv.NF_MIN = 6000
j = 0

n = len(signal)

harminv.dataCreate(n, signal, 0.0, 1.5*np.pi, nf)
harminv.solve_once()
harminv.compute_amps()

reproduction = np.zeros((len(time),), dtype=np.complex128)

for i in range(0, harminv.data.nfreqs):
    # print "Omega = ", TWOPI*get_freq(i)/dt, ", Gamma =", get_decay(i)/dt
    reproduction += (np.real(harminv.get_amplitude(i))*np.cos(harminv.TWOPI*harminv.get_freq(i)*time/dt) + 
                     np.imag(harminv.get_amplitude(i))*np.sin(harminv.TWOPI*harminv.get_freq(i)*time/dt) 
                     )*np.exp(-harminv.get_decay(i)*time/dt)

print "L1 error = ", np.linalg.norm(reproduction-signal , ord=1)/np.linalg.norm(signal, ord=1) 
print "L2 error = ", np.linalg.norm(reproduction-signal , ord=2)/np.linalg.norm(signal, ord=2) 
print "Linf error = ", np.linalg.norm(reproduction-signal , ord=np.inf)/np.linalg.norm(signal, ord=np.inf)

figure(figsize=(14,10))
plot(signal, linewidth=5)
plot(reproduction, linewidth=4)
```
